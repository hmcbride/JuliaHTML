```julia
using Winston
using Distributions
```

```julia
x=[-100:0.1:100]
@everywhere f(x)=cos(x)./(1+0.01*x.^2)
plot(x,f(x))
```

```julia
@time begin
@everywhere psize=10000
@everywhere population=rand(Distributions.Uniform(-100,100),psize)
@everywhere fitness=f(population)
@everywhere order=sortperm(fitness)
for i=1:1000
    #[population[order[i]]+=rand(Distributions.Uniform(i-(psize+1),(psize+1)-i)) for i=1:(psize-1)]
    @everywhere [population[order[j]]+=rand(Distributions.Normal(0,(psize+1-j))) for j=1:(psize-1)]
    fitness=f(population)
    order=sortperm(fitness)
end
fitness[order[psize]]
end
```
